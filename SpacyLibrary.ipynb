```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
text = nlp(u'My dream is to settle in a big firm like Amazon or Apple.I\'m working hard for these.')
```

```python
for token in text:
    print(token.text," ",token.pos_," ",token.dep_)
```

```python
for entity in text.ents:
    print(entity)
    print(entity.label_)
    print(spacy.explain(entity.label_))
```

```python
from spacy import displacy as ds                                                    
```

```python
lst=list(text.sents) #every sentence
```

```python
ds.render(lst,jupyter=True,options={"compact":True,'bg':'green','color':'#ffffff','font':'arial','distance':65},style='dep')
```

```python
ds.render(text,style='ent')
```

```python
from IPython.core.display import display, HTML

html = ds.render(text, style="dep")
display(HTML(html))                      #look in html
```
